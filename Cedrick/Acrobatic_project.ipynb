```python
class StickFigure:
  def __init__(self):
    self.x = screen_width // 2
    self.y = screen_height // 2
    self.flip = False

  def draw(self, screen):
    # Draw body
    Pygame.draw.line(screen, (0,0,0), (self.x, self.y),(self.x, self.y + 100),5)
    # Draw arms
    if self.flip:
      Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 20), (self.x - 50, self.y + 70), 5)
      Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 20), (self.x + 50, self.y + 70), 5)
    else:
      Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 20), (self.x - 50, self.y - 30), 5)
      Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 20), (self.x + 50, self.y - 30), 5)

    # Draw legs
    Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 100), (self.x - 30, self.y + 150), 5)
    Pygame.draw.line(screen, (0,0,0), (self.x, self.y + 100), (self.x +30, self.y + 150), 5)

  def flip_arms(self):
    self.flip = not self.flip
```
