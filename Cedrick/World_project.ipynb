```python
# Ensure cartopy is installed
# pip install cartopy matplotlib

import matplotlib.pyplot as plt
import cartopy.crs as ccrs

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

ax.coastlines()

ax.gridlines(draw_labels=True, linewidth=1, color='gray', alpha=0.5, linestyle='--')


plt.title('Simple World Map using Cartopy')


plt.show()

plt.figure()
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()
london_lon, london_lat = -0.1278, 51.5074
ax.plot(london_lon, london_lat, 'ro', markersize=5, transform=ccrs.PlateCarree())
ax.set_extent([-10, 5, 48, 55], crs=ccrs.PlateCarree())
ax.set_title('London on the Map')
plt.show()
```
