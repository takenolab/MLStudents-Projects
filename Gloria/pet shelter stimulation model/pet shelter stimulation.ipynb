```python
import matplotlib.pyplot as plt
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
class Pet:
    def __init__(self, pet_type, name, breed, health_status):
        self.pet_type=pet_type
        self.name = name
        self.breed = breed
        self.health_status = health_status

    def __repr__(self):
        return f'{self.pet_type}({self.name} ({self.breed}, {self.health_status}))'

class PetShelter:
    def __init__(self):
        self.pets = []

    def add_pet(self):
        pet_type=input('type the pet type')
        name = input("Enter pet name: ")
        breed = input("Enter pet breed: ")
        health_status = input("Enter pet health status (Healthy / Needs Care / Critical): ")

        if  pet_type and name and breed and health_status:
            pet = Pet(pet_type,name, breed, health_status)
            self.pets.append(pet)
            print(f"Pet '{name}' added successfully!")
        else:
            print("ERROR: Please fill all fields.")

    def remove_pet(self):
        pet_type=input('enter the pet type: ')
        name = input("Enter the name of the pet to remove: ")
        for pet in self.pets:
            if pet.name.lower() == name.lower():
                self.pets.remove(pet)
                print(f"Pet '{name}' removed successfully!")
                return
        print(f"ERROR: No pet found with name '{name}'.")

    def view_pets(self):
        if self.pets:
            print("\nPets in Shelter (sorted by health status & breed):")
            sorted_pets = sorted(self.pets, key=lambda pet: (pet.health_status, pet.breed))
            for pet in sorted_pets:
                print(f"  - {pet}")
        else:
            print("No pets in shelter.")
    def classify_pets(self):
        if not self.pets:
            print("No pets available for classification.")
            return


        df = [{"Breed": pet.breed, "Health Status": pet.health_status} for pet in self.pets]
        

        label_encoder = LabelEncoder()
        breeds = label_encoder.fit_transform([pet["Breed"] for pet in df])
        health_statuses = label_encoder.fit_transform([pet["Health Status"] for pet in df])


        X_train, X_test, y_train, y_test = train_test_split(breeds.reshape(-1, 1), health_statuses, test_size=0.3, random_state=42)

    
        model = RandomForestClassifier()
        model.fit(X_train, y_train)


        y_pred = model.predict(X_test)


        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, average='macro')
        recall = recall_score(y_test, y_pred, average='macro')
        f1 = f1_score(y_test, y_pred, average='macro')

    
        print("\nClassification Metrics:")
        print(f"Accuracy: {accuracy:.2f}")
        print(f"Precision: {precision:.2f}")
        print(f"Recall: {recall:.2f}")
        print(f"F1 Score: {f1:.2f}")

    def visualize_pets(self):
        if not self.pets:
            print("No pets to visualize.")
            return

     
        health_counts = Counter([pet.health_status for pet in self.pets])
        
      
        colors = {"Healthy": "green", "Needs Care": "yellow", "Critical": "red"}
        color_list = [colors[status] for status in health_counts.keys()]
        
        
        plt.bar(health_counts.keys(), health_counts.values(), color=color_list)
        plt.xlabel("Health Status")
        plt.ylabel("Number of Pets")
        plt.title("Pet Health Status Distribution")
        plt.show()

    def run(self):
        while True:
            print("\nPet Shelter Management")
            print("1. Add Pet")
            print("2. Remove Pet")
            print("3. View Pets")
            print("4. Visualize Pets")
            print("5. Exit")

            choice = input("Choose an option: ")

            if choice == "1":
                self.add_pet()
            elif choice == "2":
                self.remove_pet()
            elif choice == "3":
                self.view_pets()
            elif choice == "4":
                self.visualize_pets()
            elif choice == "5":
                print("Exiting program.")
                break
            else:
                print("Invalid option. Please choose again.")


shelter = PetShelter()
shelter.run()
```
