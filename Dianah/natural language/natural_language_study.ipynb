```python
from transformers import pipeline
```

```python
classfier = pipeline('sentiment-analysis')
classfier(["you really dont know what you want in life!",
           'you are too good to be here!',
           'do you really think you can pass?',
           'dude!,leave the imagination world and come back to reality.',
           'i bet you dont have sense',
           'you are pretty enough but not with a brain'])
```

```python
classfier2 =pipeline("zero-shot-classification")
classfier2(
    "life is more of the impossibles , why not take the risk today?",
    candidate_labels=["crazy friend","thoughtiful friend","spoiled friend"],
)
```

```python
generator =pipeline("text-generation")
generator("good to be true",
          max_length=12,
          num_return_sequences=3)
```

```python
name_er =pipeline('ner',grouped_entities=True)
name_er("no need for Yamie to go Must when he can go to DMI in Lilongwe")
```
